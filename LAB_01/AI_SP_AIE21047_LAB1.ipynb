```python
# Q.1:  Load the recorded speech file into your python workspace. Once loaded, plot the graph for the 
#speech signal. You may use the below code from librosa as a reference.

import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

y, rs = librosa.load('Lab_1_Ai_SP.wav')
plt.figure(figsize=(10, 5))
librosa.display.waveshow(y, color='orange')
plt.title('Graph for speech signal')


plt.axvspan(xmin=0.36 ,xmax=0.99 ,color='blue' ,alpha=0.3 ,label='AI' )
plt.axvspan(xmin=1.0 ,xmax=1.258 ,color='yellow' ,alpha=0.3 ,label='in' )
plt.axvspan(xmin=1.29 ,xmax=1.7 ,color='red' ,alpha=0.3 ,label='Speech' )
plt.axvspan(xmin=1.71 ,xmax=2.4 ,color='green' ,alpha=0.3 ,label='Processing' )

plt.legend()
plt.show()
```

```python
'''
Q.2: Observe the length & magnitude range of the signal from the plot. Observing the plot, relate to
the spoken words and silence between the words.
'''
print("Signal length:",len(y),"samples")
print("Range of Magnitude:",np.min(y),"to", np.max(y))
print("Rate of Sampling:",rs, "Hz")
```

```python
#Q.3: Take a small segment of the signal and play it. 

import IPython.display as ipd

start = int(1.43 * rs)
end = int(2.26 * rs)
segment = y[start:end]
librosa.display.waveshow(segment, color = 'orange')
ipd.Audio(segment, rate=rs)
```

```python
# EXTRA:
# SPECTOGRAM
```

```python
D = librosa.stft(y)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
S_db.shape
```

```python
# Plot the transformed audio data
fig, ax = plt.subplots(figsize=(10, 5))
img = librosa.display.specshow(S_db,
                              x_axis='time',
                              y_axis='log',
                              ax=ax)
ax.set_title('Spectogram Example', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```

```python
# MEL SPECTOGRAM:
```

```python
S = librosa.feature.melspectrogram(y=y,
                                   sr=rs,
                                   n_mels=128 * 2,)
S_db_mel = librosa.amplitude_to_db(S, ref=np.max)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
# Plot the mel spectogram
img = librosa.display.specshow(S_db_mel,
                              x_axis='time',
                              y_axis='log',
                              ax=ax)
ax.set_title('Mel Spectogram Example', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```
